# Problem 2

$$
\begin{gathered}
\min _{u(t)} x_{4}\left(t_{f}\right) \\
\text { subject to } \\
\frac{d x_{1}}{d t}=x_{2} \\
\frac{d x_{2}}{d t}=-x_{3} u+16 t-8 \\
\frac{d x_{3}}{d t}=u \\
\frac{d x_{4}}{d t}=x_{1}^{2}+x_{2}^{2}+0.0005\left(x_{2}+16 t-8-0.1 x_{3} u^{2}\right)^{2} \\
x(0)=[0 \quad-1 -\sqrt{5} \quad 0]^{T} \\
-4 \leq u \leq 10 \\
t_{f}=1
\end{gathered}
$$

```python
import numpy as np
from gekko import GEKKO

m = GEKKO()
m.time = np.linspace(0,1,101)
x1 = m.Var(value=0)
x2 = m.Var(-1)
x3 = m.Var(-np.sqrt(5))
x4 = m.Var(0)
u = m.MV(7, lb = -4, ub = 10)
u.STATUS = 1 #use for optimization
# t = m.Var(0)
t = m.Param(m.time)

mask = np.zeros(101)
mask[-1] = 1
final = m.Param(value=mask)

# m.Equation(t.dt() == 1) #why is this necessary?
m.Equation(x1.dt() == x2)
m.Equation(x2.dt() == -x3*u + 16*t - 8)
m.Equation(x3.dt() == u)
m.Equation(x4.dt() == x1**2 + x2**2 + .0005*(x2 + 16*t - 8 - 0.1*x3*u**2)**2)

m.Minimize(x4 * final)

m.options.IMODE = 6 #simultaneous control
# m.options.MV_TYPE=1

m.solve()

```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(m.time, u)

plt.figure()
plt.plot(m.time,x1, m.time,x2, m.time,x3, m.time,x4)
```

# Problem 3

$$
\begin{gathered}
\max _{u(t)} x_{2}\left(t_{f}\right) \\
\text { subject to } \\
\frac{d x_{1}}{d t}=-\left(u+0.5 u^{2}\right) x_{1} \\
\frac{d x_{2}}{d t}=u x_{1} \\
x(0)=\left[\begin{array}{ll}
1 & 0
\end{array}\right]^{T} \\
0 \leq u \leq 5 \\
t_{f}=1
\end{gathered}
$$

```python
m = GEKKO()
m.time = np.linspace(0,1,101)

x1 = m.Var(1)
x2 = m.Var(0)
u = m.MV(2.5, lb=0, ub=5)
u.STATUS = 1

mask = np.zeros(101)
mask[-1] = 1
final = m.Param(mask)

m.Equation(x1.dt() == -(u + 0.5*u**2)*x1)
m.Equation(x2.dt() == u*x1)

m.Maximize(x2 * final)

m.options.IMODE = 6

m.solve()

plt.plot(m.time, u)

plt.figure()
plt.plot(m.time, x1, m.time, x2)

```

# Minimize Final Time (Jennings)

[Link](https://apmonitor.com/do/index.php/Main/MinimizeFinalTime) for this problem.

$$
\begin{gathered}
\min _{u(t)} t_{f} \\
\text { subject to } \\
\frac{d x_{1}}{d t}=t_{f} u \\
\frac{d x_{2}}{d t}=t_{f} \cos \left(x_{1}(t)\right) \\
\frac{d x_{3}}{d t}=t_{f} \sin \left(x_{1}(t)\right) \\
x(0)=[\pi / 2,4,0] \\
x_{2}\left(t_{f}\right)=0 \\
x_{3}\left(t_{f}\right)=0 \\
-2 \leq u(t) \leq 2
\end{gathered}
$$

```python
m = GEKKO()

m.time = np.linspace(0,1,101)
tf = m.FV(1, .001, 100)
x1 = m.Var(np.pi/2)
x2 = m.Var(4)
x3 = m.Var(0)
u = m.MV(0, lb = -2, ub=2)
u.STATUS = 1

mask = np.zeros(101)
mask[-1] = 1
final = m.Param(value=mask)

m.Equation(x1.dt() == tf*u)
m.Equation(x2.dt() == tf*m.cos(x1))
m.Equation(x3.dt() == tf*m.sin(x1))
m.Equation(x2 * final == 0)
m.Equation(x3*final == 0)

m.Minimize(tf)

m.options.IMODE = 6

m.solve()
```

# Example 3

$$
\begin{gathered}
\max _{u(t)} x_{2}\left(t_{f}\right) \\
\text { subject to } \\
\frac{d x_{1}}{d t}=-\left(u+0.5 u^{2}\right) x_{1} \\
\frac{d x_{2}}{d t}=u x_{1} \\
x(0)=[10]^{T} \\
0 \leq u \leq 5 \\
t_{f}=1
\end{gathered}
$$

```python
m = GEKKO()

m.time = np.linspace(0,1,101)
x1 = m.Var(1)
x2 = m.Var(0)
u = m.Var(2.5,0,5)


mask = np.zeros(101)
mask[-1] = 1
final = m.Param(value=mask)

m.Equation(x1.dt() == -(u + 0.5*u**2)*x1)
m.Equation(x2.dt() == u*x1)

m.options.IMODE = 6

m.Maximize(x2 * final)

m.solve()
```

```python
plt.plot(m.time, x1, m.time, x2, m.time, u)
```

# Example 4

$$
\begin{gathered}
\max _{T(t)} x_{2}\left(t_{f}\right) \\
\text { subject to } \\
\frac{d x_{1}}{d t}=-k_{1} x_{1}^{2} \\
\frac{d x_{2}}{d t}=k_{1} x_{1}^{2}-k_{2} x_{2} \\
k_{1}=4000 \exp \left(-\frac{2500}{T}\right) \\
k_{2}=6.2 e 5 \exp \left(-\frac{5000}{T}\right) \\
x(0)=[1,0]^{T} \\
298 \leq T \leq 398 \\
t_{f}=1
\end{gathered}
$$

```python
m = GEKKO()

m.time = np.linspace(0,1,101)
x1 = m.Var(1)
x2 = m.Var(0)
T = m.Var(300, 298, 398)
k1 = m.Intermediate(4000*m.exp(-2500/T))
k2 = m.Intermediate(6.2e5*m.exp(-5000/T))

mask = np.zeros(101)
mask[-1] = 1
final = m.Param(value=mask)
m.Equation(x1.dt() == -k1*x1**2)
m.Equation(x2.dt() == k1*x1**2 - k2*x2)
m.Maximize(x2*final)

m.options.IMODE = 6

m.solve()
```

```python
plt.plot(m.time,x1, m.time,x2)
plt.figure()
plt.plot(m.time,T)
```

```python
from gekko import GEKKO
import numpy as np


m = GEKKO()
m.time = np.linspace(0,1,101)
u = m.MV(value=.75) #mv is what you use as a DOF for optimization problem. 
u.STATUS = 1
x1 = m.Var(value=1)
x2 = m.Var(value=0)

m.Equation(x1.dt() >= u)
m.Equation(x2.dt() == x1**2 + u**2)

#multiply all x2 trajectory by zero excepto for last one. 
p = np.zeros(101)
p[-1] = 1
final = m.Param(p)

m.Minimize(x2*final)

m.options.IMODE = 6

m.solve()
```

```python
import matplotlib.pyplot as plt
plt.plot(m.time, u, m.time, x1, m.time, x2)
```
